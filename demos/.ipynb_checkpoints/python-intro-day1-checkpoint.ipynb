Assignment Dynamics
==

```python
A = 5  #Assign the object "5" to the name A
B = A  #Assign the object that A referes to to the name B
print("A=",A,"B=",B 
```

```python
A = 7 #Assign the object "7" to the name A              
print("A=",A,"B=",B 
```

```python
A = [1,2,3,4,5] #Assign a list object to the name A
B = A           #Assign the object that A referes to to the name B 
print("A=",A,"B=",B 
```

```python
A = [1,2,3,4,5] #Assign a list object to the name A
B = A           #Assign the object that A referes to to the name B
A = [5,4,3,2,1] #Assign a different list object to the name A       
print("A=",A,"B=",B  
```

```python
A = [1,2,3,4,5] #Assign a list object to the name A
B=A             #Assign the object that A referes to to the name B
B.append(6)     #Question: What if we append an item to the list object that B refers to?  
print("A=",A,"B=",B 
```

```python
A = [1,2,3,4,5] #Assign a list object to the name A
B = A           #Assign the object that A referes to to the name B
B[0]=-1         #Question: What if we change an item in the list object that B refers to?  
print("A=",A,"B=",B
```

If-Elif-Else

```python
#Basic If-Elif-Else
X=5
if(X<0):
    print("X is negative")
elif(X==0):
    print("X is zero")
else:
    print("X is positive")
```

Functions
==

```python
def f(a):             #Define a basic function of 1-argument
    return a**2

print("f(2)=",f(2)
```

```python
def g(h,a,b):         #Define a function of 3-arguments where h is a function and a and b are numbers
    return h(a)+h(b)

print("g(f,2,3)=",g(f,2,3)
```
